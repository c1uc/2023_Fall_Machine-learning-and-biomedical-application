## Homework 4: Implement the LDA using sklearn

```python
import matplotlib.pyplot as plt
from sklearn import datasets

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

data = datasets.load_iris()
X, y = data.data, data.target

# You need to finish the following codes to display the result
clf = LDA(n_components=2)
X_ = clf.fit_transform(X, y)

plt.scatter(X_[:, 0], X_[:, 1], c=y, edgecolor="none", alpha=0.8, cmap=plt.get_cmap("viridis", 3))

plt.xlabel("Linear Discriminant 1")
plt.ylabel("Linear Discriminant 2")
plt.colorbar()
plt.show()
```
